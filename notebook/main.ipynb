```python
import pandas as pd
```

```python
land_df = pd.read_csv('../data/landing.csv', encoding='SHIFT_JIS')
land_df.columns = ['year', *['m{}'.format(i) for i in range(1, 13)], 'count']
```

```python
land_df.head()
```

```python
land_df.describe()
```

```python
land_df.shape[0] - land_df.isnull().sum()
```

```python
# 1 ~ 5, 11, 12月は台風の上陸数が少ないので削除
use_months = ['year', 'count', 'm6', 'm7', 'm8', 'm9', 'm10']
land_df = land_df[use_months]
```

```python
land_df.head()
```

```python
land_df.fillna(0, inplace=True)
```

```python
# 1982年からしか月平均海面水温のデータがないので、それ以外を削除
land_df = land_df[land_df['year'] >= 1982]
```
